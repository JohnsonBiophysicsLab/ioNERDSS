```python
from math import *
import numpy as np

def rot(coord, angx, angy, angz):
    x = coord[0]
    y = coord[1]
    z = coord[2]
    x_ = x*(cos(angy)*cos(angz)-sin(angx)*sin(angy)*sin(angz))-y*cos(angx)*sin(angz)+z*(sin(angy)*cos(angz)+sin(angx)*cos(angy)*sin(angz))
    y_ = x*(cos(angy)*sin(angz)+sin(angx)*sin(angy)*cos(angz))+y*cos(angx)*cos(angz)+z*(sin(angy)*sin(angz)-sin(angx)*cos(angy)*cos(angz))
    z_ = x*(-cos(angx)*sin(angy))+y*sin(angx)+z*cos(angx)*cos(angy)
    result = [round(x_, 8), round(y_, 8), round(z_, 8)]
    return np.array(result)

rot([0, 0, 0], 2, 3, 4)
```

```python
from mpl_toolkits.mplot3d import axes3d  
import matplotlib.pyplot as plt  

def sig_vis2(list, ax, n):
    for i in range(1, 4):
        figure = ax.plot([list[0][0], list[i][0]], [list[0][1], list[i][1]], [list[0][2], list[i][2]])
    figure = ax.plot([n[0], list[0][0]], [n[1], list[0][1]], [n[2], list[0][2]])
        
def vis2(lst, n):
    fig = plt.figure(1)  
    ax = fig.gca(projection='3d')
    for element in lst:
        sig_vis2(element, ax, n) 
    plt.show()
```

```python
COM =  [0, 0, 0]
lg1 =  [-11.99744871, -6.92673024, -19.59175171]
lg2 =  [11.99744871, -6.92673024, -19.59175171]
lg3 =  [0, 13.85346049, -19.59175171]
n =  [0, 0, 1]
vis2([[COM, lg1, lg2, lg3]], n)
```

```python
COM_ = rot(COM, 2, 3, 4)
lg1_ = rot(lg1, 2, 3, 4)
lg2_ = rot(lg2, 2, 3, 4)
lg3_ = rot(lg3, 2, 3, 4)
n_ = rot(n, 2, 3, 4)
vis2([[COM_, lg1_, lg2_, lg3_]], n_)
```
